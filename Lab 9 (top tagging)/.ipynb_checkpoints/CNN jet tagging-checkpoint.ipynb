```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import pandas as pd
from tqdm import tqdm
import os
```

# Load Data

```python
from numpy import load

directory = 'jets/jets'
labels = ['Gluon', 'Light quark', 'W boson', 'Z boson', 'Top quark']
features = ['n', 'phi', 'E', 'R', 'length','label']

nf = 0
cap = 10000

df = pd.DataFrame(columns = features)
nf = 0
for filename in tqdm(os.scandir(directory)):
    if (nf == cap):
        break
    nf += 1
    with np.load(filename) as npz:
        flist = npz['x'].T
        llist = npz['y']
        lb = labels[np.where(llist == 1)[0][0]]
        tmp = pd.DataFrame({'n':[flist[1]], 'phi':[flist[2]], 'E':[flist[3]], 'R':[flist[4]],'length': len(npz['x']), 'label':lb})
        df = df.append(tmp, ignore_index = True)
                

#X = df.loc[:, df.columns != 'label']
#y = df.label
```

# Preprocessing

```python
import torch
from tqdm import tqdm


from sklearn.preprocessing import MinMaxScaler

ds_size = 200000

le = 16
wi = 16

X = []
y = np.array([])

labels = ['Gluon', 'Light quark', 'W boson', 'Z boson', 'Top quark']
        

def normalize(arr, t_min, t_max):
    norm_arr = []
    diff = t_max - t_min
    minarr = -1
    maxarr = 1
    diff_arr = maxarr - minarr
    for i in arr:
        temp = (((i - minarr)*diff)/diff_arr) + t_min
        norm_arr.append(temp)
    return norm_arr

# Change the normalizer so that 0 is 8, -1 is 0, 16 is 1
# set ranges to from 0-16 and -1-1


for x in tqdm(range(ds_size)):
    v1 = df.iloc[x]['n']
    v2 = df.iloc[x]['phi']   
    w = np.round(normalize(v1, 0, wi-1), 0).astype(int)
    h = np.round(normalize(v2, 0, le-1), 0).astype(int)
    img = np.zeros((le, wi))
    la = labels.index(df.iloc[x]['label'])
    for i in range(0, len(w)-1):
        arr = df.iloc[x]['E']
        img[w[i]-1, h[i]-1] += arr[i]
    
    X.insert(x, img)
    y = np.append(y, la)
   

X = np.asarray(X)
print(X.shape)
```

# Model

```python
import torch.nn as nn

class CNNModel(torch.nn.Module):
    
    def __init__(self):
        
        super(CNNModel, self).__init__()
        
        self.cnn1 = nn.Conv2d(in_channels=1, out_channels=8, kernel_size=5)
        self.cnn2 = nn.Conv2d(in_channels=8, out_channels=16, kernel_size=5)
        self.cnn3 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=5)
        self.maxpool1 = nn.MaxPool2d(kernel_size=2)
        self.maxpool2 = nn.MaxPool2d(kernel_size=2)
        self.maxpool3 = nn.MaxPool2d(kernel_size=2)
        self.fc1 = nn.Linear(32 , 100)
        #self.fc2 = nn.Linear(100, 50)
        #self.fc3 = nn.Linear(50, 25)

    def forward(self, x):
        
        #x = x.permute(0, 2, 1, 3)
        x = nn.functional.relu(self.cnn1(x))
        x = self.maxpool1(x)
        x = nn.functional.relu(self.cnn2(x))    
        x = self.maxpool2(x)
        x = nn.functional.relu(self.cnn3(x))
        #x = self.maxpool3(x)
        x = x.reshape(x.shape[0], -1)
        #x = nn.functional.relu(self.fc1(x))
        #x = nn.functional.relu(self.fc2(x))
        x = self.fc1(x)
        
        return x
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X.reshape((ds_size, wi * le))).reshape((ds_size, wi, le))

#split up data

ntrain = ds_size*0.8

X_train = X[:int(ntrain)]
y_train = y[:int(ntrain)]

X_test = X[int(ntrain):]
y_test = y[int(ntrain):]

X_val = X_train[:(int(ntrain*0.1))]
y_val = y_train[:(int(ntrain*0.1))]
torch.manual_seed(55)

X_train = np.reshape(X_train, (int(ntrain), 1, le, wi))
X_val = np.reshape(X_val, (int(ntrain*0.1), 1, le, wi))
X_test = np.reshape(X_test, (int(ds_size - ntrain), 1, le, wi))

model = CNNModel()
learning_rate = 0.0005
epochs = 15
batchsize = 100
loss_func = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = learning_rate)

model
```

```python
train_loss_list = []
validation_accuracy_list = np.zeros((epochs,))
training_accuracy_list = np.zeros((epochs,))
```

```python
## WITH BATCHES

import tqdm
# Features -> floating numbers, Target labels -> int64 format
X_train = torch.from_numpy(X_train).float()
y_train = torch.from_numpy(y_train).long()

#input = input.permute(0, 2, 1, 3)

X_val = torch.from_numpy(X_val).float()
y_val = torch.from_numpy(y_val).long()

X_test = torch.from_numpy(X_test).float()
y_test = torch.from_numpy(y_test).long()

batch_X = torch.split(X_train, batchsize)
batch_y = torch.split(y_train, batchsize)
batch_split = len(batch_X)

for epoch in tqdm.trange((epochs)):
    
    batch_acc = []
    
    for k in range(batch_split): 
        optimizer.zero_grad()
        #print(train_batches_features[k].size())
        batch_output = model(batch_X[k])
        loss = loss_func(batch_output, batch_y[k])
        train_loss_list.append(loss.item())
        loss.backward()
        optimizer.step()
        
        #tr_cor = (torch.argmax(batch_output, dim=1) == batch_y).type(torch.FloatTensor)
        #batch_acc.append(tr_cor.mean())
    
    #avg_batch_acc = sum(batch_acc)/len(batch_acc)
    avg_batch_acc = 0
    #training_accuracy_list[epoch] = avg_batch_acc
    
    with torch.no_grad():

        val_outputs = model(X_val)
        val_cor = (torch.argmax(val_outputs, dim=1) == y_val).type(torch.FloatTensor)
        val_per = np.round(val_cor.mean().numpy() * 100, 2)
        validation_accuracy_list[epoch] = val_cor.mean()
        
        print("epoch: "+ str(epoch), "tr_acc: " + str(avg_batch_acc) + '%', "val_acc: " + str(val_per) + '%', "training loss: " + str(loss.item()), flush=True)
```

```python
with torch.no_grad():
    
    y_pred_test = model(X_test)
    
    correct = (torch.argmax(y_pred_test, dim=1) == y_test).type(torch.FloatTensor)
    
    print("Testing Accuracy: " + str(correct.mean().numpy()*100))
    
```
